```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
#%matplotlib qt
```

```python
#reading and showing an image
image = mpimg.imread("perspective transform.jpg")
plt.imshow(image)
plt.show()
```

```python
plt.imshow(image)
plt.plot(165,55,'.',COLOR='black') #top left
plt.plot(165,88,'.',COLOR='black') #bottom left
plt.plot(235,55,'.',COLOR='black') #top right
plt.plot(235,88,'.',COLOR='black')  #bottom right
plt.show()
```

```python
#define perspective transfoorm
def wrap(img):
    image = np.array(img)
    size_img = (image.shape[1],image.shape[0])
    #print(size_img)
    #source coordinates
    src = np.float32(
                     [[165,55],
                      [165,88],
                      [235,55],
                      [235,88]])
    #destiation coordinates
    des = np.float32(
                      [[75,55],
                      [75,88],
                      [235,55],
                      [235,88]])
    
    #compute the perspective matrix
    M = cv2.getPerspectiveTransform(src,des)
    
    wraped = cv2.warpPerspective(img,M,size_img,flags=cv2.INTER_LINEAR)
    return wraped
```

```python
%matplotlib inline
wrapped_img = wrap(image)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=30)
ax2.imshow(wrapped_img)
ax2.set_title('New Perspective', fontsize=30)
plt.show()
```
